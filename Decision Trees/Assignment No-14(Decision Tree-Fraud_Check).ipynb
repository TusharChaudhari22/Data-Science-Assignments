```python
#Treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

Fraud = []

for i in data['Taxable.Income']:
    if i <= 30000:
        Fraud.append('Risky')
    else:
        Fraud.append('Good')
```

```python
Fraud
```

```python
data['Fraud'] = Fraud
```

```python
data
```

```python
data[data['Taxable.Income'] <= 30000]
```

```python
#Ordinal encoading for independent variables.

oe = OrdinalEncoder()

data[['Undergrad','Marital.Status','Urban']] = oe.fit_transform(data[['Undergrad','Marital.Status','Urban']])
```

```python
data
```

#### Model Building

```python
x = data.iloc[:,0:-1]
y = data['Fraud']
```

```python
#Lebal encoading for target variable.

le = LabelEncoder()

le.fit_transform(y)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

##### Using = C5.0

```python
dt = DecisionTreeClassifier(criterion='entropy')

dt.fit(xtrain,ytrain)

ypred = dt.predict(xtest)

print(classification_report(ytest,ypred))
```

```python
print('Training Accuracy = ', dt.score(xtrain,ytrain))
print('Testing Accuracy = ', dt.score(xtest,ytest))
```

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.rcParams['figure.dpi'] = 250
```

```python
#Decision tree.

tree.plot_tree(dt, feature_names=x.columns, class_names=data['Fraud'].unique(), filled=True)
plt.show()
```

##### Using = CART - Gini Impurities

```python
dt1 = DecisionTreeClassifier(criterion='gini')

dt1.fit(xtrain,ytrain)

yp = dt1.predict(xtest)

print(classification_report(ytest,yp))
```

```python
print('Training accuracy = ', dt1.score(xtrain,ytrain))
print('Testing Accuracy = ', dt1.score(xtest,ytest))
```

```python
#Decision tree.

tree.plot_tree(dt1, feature_names=x.columns, class_names=data['Fraud'].unique(), filled=True)
plt.show()
```
