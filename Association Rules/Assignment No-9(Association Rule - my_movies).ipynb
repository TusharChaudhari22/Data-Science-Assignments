```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read my_movies Dataset
movies = pd.read_csv("my_movies.csv")
movies.head()
```

```python
movies.shape
```

```python
#Apply Filtering
movies = movies.iloc[:,5:]
movies
```

##### Association Rule

```python
#Libraries
from mlxtend.frequent_patterns import apriori, association_rules

import warnings
warnings.filterwarnings('ignore')
```

###### Rules with min_support >= 0.1

```python
#Rules
score1 = apriori(movies, min_support=0.1, use_colnames=True, max_len=3)
score1['Length'] = score1['itemsets'].apply(lambda x : len(x))
score1
```

```python
# Slicing using support value and length.

# Rule with supprt value >= 0.5 & length = 2
score1[(score1['support'] >= 0.5) & (score1['Length'] == 2)]
```

```python
# Slicing using support value and length.

# Rule with support value = 0.4 & length = 3
score1[(score1['support'] == 0.4) & (score1['Length'] == 3)]
```

```python
#model building
model1 = association_rules(score1, metric='lift', min_threshold=1)
model1
```

```python
#Slicing using support value and confidence value

#Rules with support >= 0.5 and confidence >= 0.7
model1[(model1['support'] >= 0.5) & (model1['confidence'] >= 0.7)]
```

```python
#Distribution of data
model1[['support','confidence','lift']].hist()
plt.tight_layout()
```

```python
#Relation between support and confidence.
sns.scatterplot(model1['support'], model1['confidence']);
plt.title("Support v/s Confidence");
```
